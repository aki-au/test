```python
import os
import random
import shutil


# Set paths
base_dir = "/content/data"
output_dir = "/content/randomized_data"

subfolders = ["val", "dev", "test"]


all_subjects = set()
for subfolder in subfolders:
    folder_path = os.path.join(base_dir, subfolder)
    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            subject = "_".join(file.split("_")[:-1])
            all_subjects.add(subject)


random_subjects = random.sample(sorted(all_subjects), 20)



for subfolder in subfolders:
    os.makedirs(os.path.join(output_dir, subfolder), exist_ok=True)


for subject in random_subjects:
    for subfolder in subfolders:
        src_path = os.path.join(base_dir, subfolder)
        dst_path = os.path.join(output_dir, subfolder)
        filename = f"{subject}_{subfolder}.csv"
        src_file = os.path.join(src_path, filename)
        if os.path.exists(src_file):
            shutil.copy(src_file, os.path.join(dst_path, filename))


random_subjects
```
